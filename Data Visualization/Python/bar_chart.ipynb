```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
#Running this line it wil enhance the visual when using Matplotlib
sns.set()

#Get the data
df_used_cars = pd.read_csv("../Datasets/bar_chart_data.csv")

#Matplotlib
plt.figure(figsize = (9,6))
plt.bar(x = df_used_cars["Brand"], 
        height = df_used_cars["Cars Listings"],
        color = "midnightblue")
plt.xticks(rotation = 45, fontsize = 13)
plt.ylabel("Number of Listings", fontsize = 13)
plt.title("Cars Listing by Brand", fontsize=16, fontweight="bold")
plt.show()

#To save the plot as image
plt.savefig("used_cars_bars.png")
```
